#### Run this cell to connect to your GIS and get started:

```python
from arcgis.gis import GIS
gis = GIS("home")
%autosave 60
```

#### Display  never logged in users (never logged in and created over two years ago) and inactive users (have logged in but not for over 2 years)

```python
import time
count = 0
for user in gis.users.search(sort_field='created', sort_order='asc', max_users=600):
    
   
       
    # convert Unix epoch time to local time
    created_time = time.localtime(user.created/1000)
    last_accessed = time.localtime(user.lastLogin/1000)
    #print (last_accessed[0])
        
        
    if str(last_accessed[0]) == '1969'and str(created_time[0]) < '2022':
        print(user.username + ' (never logged in)')
        print("Created: {}/{}/{}".format(created_time[0], created_time[1], created_time[2]))
        print("Last active: {}/{}/{}".format(last_accessed[0], last_accessed[1], last_accessed[2]))
        count = count + 1
        
    #print('inactive')
        
    
    if str(last_accessed[0]) > '1969' and str(last_accessed[0]) < '2022':
        print(user.username + ' (inactive)')
        print("Created: {}/{}/{}".format(created_time[0], created_time[1], created_time[2]))
        print("Last active: {}/{}/{}".format(last_accessed[0], last_accessed[1], last_accessed[2]))
        count = count + 1
      
    
print(str(count) + ' inactive users')              
                
```
